```python
# pd.read_csv('2110_0.csv')
# pular as duas primeiras linhas do arquivvo pois não nos interessa ( skiprows)
df = pd.read_excel('2120_0.xlsx', skiprows=2)
df.head()
```

```python
# vamos gerar novamente o dataframe descartando as ultimas 16 linhas ( skipfooter)
df = pd.read_excel('2120_0.xlsx', skiprows=2 , skipfooter=16)
```

```python
# não sera exibido as linhas faltantes
df.tail()
```

```python
# verificar todos os tipos das colunas
df.dtypes
```

```python
# Data de lançamento deveria estar como data e não object
# converter a coluna Data de lançamento para tipo data
# colcjetes sempre significa selecionar
df['Data de lançamento'] = pd.to_datetime(df['Data de lançamento'])
```

```python
# identificar o problema da coluna Data de lançamento
# 175 elemntos com o nome relaçamento na coluna Data de lançamento, onde o esperado era data e não texto
df['Data de lançamento'].value_counts()
```

```python
# identificcando alguns files que estão com data = relançamento
df[df['Data de lançamento'] == 'Relançamento'].head()
```

### Decisão sobre dados incorretos/faltantes/divergentes

```python
df.shape
```

```python
# criar novo dataframe para preservar o original, e sem os relançamentos
# df_novo , pega todos os valores que estejam em Data de lançamento e sejam diferentes de Relançamento
# copy() cria uma copia identica ao df e não uma referencia em memoria
df_novo = df[df['Data de lançamento'] != 'Relançamento'].copy()
```

```python
df_novo['Data de lançamento'].value_counts()
# temos agora so campo com numeros
```

```python
df_novo.dtypes
# Data de lançamento -> object
```

```python
df_novo['Data de lançamento'] = pd.to_datetime(df_novo['Data de lançamento'])
```

```python
df_novo.head()
```

```python
df_novo.dtypes
# Data de lançamento -> datetime64
```

```python
df_novo.shape
# foram excluidas linhas, original era 5691 e o atual é 5516
```
