```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
salario = np.array([100,200,300,500,400,150])
```

```python
salario
```

```python
plt.plot(salario)
```

```python
plt.plot(salario, c='Black') # c colore a linha
```

```python
plt.plot(salario, c='Orange')
```

```python
plt.plot(salario, c='Black', ls='--') # ls cria linha tracejada
```

```python
plt.plot(salario, c='Black', ls='--', marker='s') # marker adiciona marcadores
```

```python
plt.plot(salario, c='Black', ls='--', marker='s', ms='18') # ms aumenta o tamanho do marker
```

```python
salarios_marcos = np.array([100,200,400,300,500,600,1000])
salarios_gileno = np.array([750,300,500,450,200,450,800])
salarios_pedro = np.array([500,350,600,700,900,2000,1100])
salarios_janderson = np.array([1200,800,900,1500,3000,1900,2200])
```

```python
plt.plot(salarios_marcos, c='Black',marker='s', ls='--', ms=10, label='Salario Marcos')
plt.plot(salarios_gileno, c='Red',marker='^', ls='--', ms=10, label='Salario Gileno')
plt.plot(salarios_pedro, c='Green',marker='o', ls='--', ms=10, label='Salario Pedro')
plt.plot(salarios_janderson, c='Orange',marker='D', ls='--', ms=10, label='Salario Janderson')
plt.legend() # colocando legenda no plot
plt.show()
```
