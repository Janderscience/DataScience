##  comprimento da sepala

```python
plt.plot(iris_setosa, c="Red", ls=":", marker="s", ms=8, label="Iris-Setosa")
plt.plot(iris_versicolor,c="Green",ls="--",marker="D",ms=7, label="Iris-Versicolor")
plt.plot(iris_virginica,c="Blue",ls="--",marker="o",ms=7, label="Iris-Virginica")
plt.title("Comprimento da Sépala",fontsize=27)
plt.legend()
plt.show()
```

## Largura da Sépalas
