```python
esquerda = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
direita = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
pd.merge(esquerda, direita, on=['key1', 'key2'])
```

```python
esquerda = pd.DataFrame({'localidade':['MG', 'SP', 'RJ', 'DF'],
                        'brancos': ['345', '300', '690', '453'],
                        'nulos': ['567', '216', '57', '789']})
direita = pd.DataFrame({'localidade': ['MG', 'SP', 'RJ', 'DF'],
                          'não opniram': ['457', '211', '685', '643'],
                          'votantes': ['386', '128', '678', '529']})
```

```python
esquerda
```

```python
direita
```

```python
pd.merge(esquerda,direita,how='inner',on='localidade')
```

```python
pd.merge(esquerda,direita,how='outer',on='localidade')
```

```python
esquerda.join(direita, how='outer',rsuffix='-')
```

```python
esquerda.join(direita, how='inner',rsuffix='-')
```

```python
esquerda.join(direita, how='left',rsuffix='-')
```

```python
esquerda.join(direita, how='right',rsuffix='-')
```
